```python
import requests
from bs4 import BeautifulSoup
import re

def retrieve(string):
  temp = ""
  flag = True
  while (string[0] in ("\t", " ", " ", " ", "\n")):
    string = string[1: ]

  for n in string:
    if (flag):
      if (n == "<"):
        flag = False
        continue
      temp += n
    elif (n == ">"):
      flag = True

  return temp

def get_page():
  try:
    resp = requests.get("https://cjwu.github.io/list.html")
  except:
    resp = None

  if (resp and resp.status_code == 200):
    soup = BeautifulSoup(resp.text, 'html.parser')
    li_list = soup.find_all("li")
    pattern = re.compile('<.*?>')
    for li in li_list:
      print(retrieve(str(li)))

get_page()
```
