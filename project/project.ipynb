# 爬蟲程式

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
from selenium import webdriver

def find(url):
    result = []
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, "html.parser")
    poem = soup.find_all("a", {"target": "_blank"})
    
    for p in poem.copy():
        if(not p["href"][1].isnumeric()):
            poem.remove(p)
    
    ua = UserAgent()
    option = webdriver.ChromeOptions()
    option.add_argument("headless")
    for p in poem:
        user = ua.random
        option.add_argument("user-agent=" + user)
        driver = webdriver.Chrome(options = option)
        
        driver.get("http://m.shangshiwen.com"+ p["href"])
        page_source = driver.page_source
        
        soup2 = BeautifulSoup(page_source, "html.parser")
        content = soup2.find("div", {"class": "shicineirong", "id": "shicineirong"})
        try:
            result.append(content.text)
        except:
            pass
    
    return result
```

# 資料預處理

```python
from nltk.corpus import stopwords
import string
import jieba

def tokenize_sentence(sentence):
    stop_words = stopwords.words("chinese")
    stop_words2 = ("\n", "？", "：", "\t", "，", "。", "；", " ", "！", "¤")

    tokens = jieba.cut(sentence, cut_all = False, HMM = True)
    tokens = [i for i in tokens if i not in string.punctuation]
    tokens = [i for i in tokens if i not in stop_words]
    tokens = [i for i in tokens if i not in stop_words2]

    return tokens
```

```python
tokens1, tokens2, tokens3, tokens4, tokens5 = [], [], [], [], []

for poem in temp1:
    tokens1.append(tokenize_sentence(poem))
for poem in temp2:
    tokens2.append(tokenize_sentence(poem))
for poem in temp3:
    tokens3.append(tokenize_sentence(poem))
for poem in temp4:
    tokens4.append(tokenize_sentence(poem))
for poem in temp5:
    tokens5.append(tokenize_sentence(poem))
```

```python
import pandas as pd
from collections import Counter

temp = []
for i, tokens in enumerate(tokens1):
    dic = {}
    dic["author"] = "李白"
    dic.update(Counter(tokens))
    info = pd.DataFrame(dic, index = [i])
    temp.append(info)
for i, tokens in enumerate(tokens2):
    dic = {}
    dic["author"] = "杜甫"
    dic.update(Counter(tokens))
    info = pd.DataFrame(dic, index = [i])
    temp.append(info)
for i, tokens in enumerate(tokens3):
    dic = {}
    dic["author"] = "白居易"
    dic.update(Counter(tokens))
    info = pd.DataFrame(dic, index = [i])
    temp.append(info)
for i, tokens in enumerate(tokens4):
    dic = {}
    dic["author"] = "蘇軾"
    dic.update(Counter(tokens))
    info = pd.DataFrame(dic, index = [i])
    temp.append(info)
for i, tokens in enumerate(tokens5):
    dic = {}
    dic["author"] = "辛棄疾"
    dic.update(Counter(tokens))
    info = pd.DataFrame(dic, index = [i])
    temp.append(info)

data = pd.concat(temp, axis = 0)
```

# tf-idf

```python
import collections

record = set()
idf_count = collections.defaultdict(int)
lengthOfArcticles = 0

for poem in temp1:
    tokens = tokenize_sentence(poem)
    for token in set(tokens):
        idf_count[token] += 1
    lengthOfArcticles += 1
for poem in temp2:
    tokens = tokenize_sentence(poem)
    for token in set(tokens):
        idf_count[token] += 1
    lengthOfArcticles += 1
for poem in temp3:
    tokens = tokenize_sentence(poem)
    for token in set(tokens):
        idf_count[token] += 1
    lengthOfArcticles += 1
for poem in temp4:
    tokens = tokenize_sentence(poem)
    for token in set(tokens):
        idf_count[token] += 1
    lengthOfArcticles += 1
for poem in temp5:
    tokens = tokenize_sentence(poem)
    for token in set(tokens):
        idf_count[token] += 1
    lengthOfArcticles += 1
```

```python
import pandas as pd
import math

x_data, y_data = [], []
for poem in temp1:
    chart = collections.Counter(tokenize_sentence(poem))
    record = {}
    for w, n in chart.items():
        tf = round(n / sum(chart.values()), 4)
        idf = round(lengthOfArcticles / idf_count[w], 4)
        record[w] = round(tf * math.log(idf, 10), 4)
        
    info_x = pd.DataFrame(record, index = [len(x_data)])
    x_data.append(info_x)
    
    info_y = pd.DataFrame({"author": "李白"}, index = [len(y_data)])
    y_data.append(info_y)
    
for poem in temp2:
    chart = collections.Counter(tokenize_sentence(poem))
    record = {}
    for w, n in chart.items():
        tf = round(n / sum(chart.values()), 4)
        idf = round(lengthOfArcticles / idf_count[w], 4)
        record[w] = round(tf * math.log(idf, 10), 4)
        
    info_x = pd.DataFrame(record, index = [len(x_data)])
    x_data.append(info_x)
    
    info_y = pd.DataFrame({"author": "杜甫"}, index = [len(y_data)])
    y_data.append(info_y)
    
for poem in temp3:
    chart = collections.Counter(tokenize_sentence(poem))
    record = {}
    for w, n in chart.items():
        tf = round(n / sum(chart.values()), 4)
        idf = round(lengthOfArcticles / idf_count[w], 4)
        record[w] = round(tf * math.log(idf, 10), 4)
        
    info_x = pd.DataFrame(record, index = [len(x_data)])
    x_data.append(info_x)
    
    info_y = pd.DataFrame({"author": "白居易"}, index = [len(y_data)])
    y_data.append(info_y)
    
for poem in temp4:
    chart = collections.Counter(tokenize_sentence(poem))
    record = {}
    for w, n in chart.items():
        tf = round(n / sum(chart.values()), 4)
        idf = round(lengthOfArcticles / idf_count[w], 4)
        record[w] = round(tf * math.log(idf, 10), 4)
        
    info_x = pd.DataFrame(record, index = [len(x_data)])
    x_data.append(info_x)
    
    info_y = pd.DataFrame({"author": "蘇軾"}, index = [len(y_data)])
    y_data.append(info_y)
    
for poem in temp5:
    chart = collections.Counter(tokenize_sentence(poem))
    record = {}
    for w, n in chart.items():
        tf = round(n / sum(chart.values()), 4)
        idf = round(lengthOfArcticles / idf_count[w], 4)
        record[w] = round(tf * math.log(idf, 10), 4)
        
    info_x = pd.DataFrame(record, index = [len(x_data)])
    x_data.append(info_x)
    
    info_y = pd.DataFrame({"author": "辛棄疾"}, index = [len(y_data)])
    y_data.append(info_y)
        
print("Concating x...")
x = pd.concat(x_data, axis = 0)
x = x.fillna(0)
print("Concating y...")
y = pd.concat(y_data, axis = 0)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, stratify = y)
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier()
print("Training KNN...")
KNN.fit(x_train, y_train)
print("Predicting KNN...")
prediction = KNN.predict(x_test)
```

```python
result = y_test["author"].values
```

```python
n = 0

for i in range(len(prediction)):
    if (prediction[i] == result[i]):
        n += 1

print(f"Accuracy: {n / len(prediction) * 100}%")
```

# 調整

```python
from collections import defaultdict

a1, a2, a3, a4, a5 = defaultdict(int), defaultdict(int), defaultdict(int), defaultdict(int), defaultdict(int)

for poem in temp1:
    tokens = tokenize_sentence(poem)
    for token in tokens:
        a1[token] += 1
for poem in temp2:
    tokens = tokenize_sentence(poem)
    for token in tokens:
        a2[token] += 1
for poem in temp3:
    tokens = tokenize_sentence(poem)
    for token in tokens:
        a3[token] += 1
for poem in temp4:
    tokens = tokenize_sentence(poem)
    for token in tokens:
        a4[token] += 1
for poem in temp5:
    tokens = tokenize_sentence(poem)
    for token in tokens:
        a5[token] += 1
```

```python
author1, author2, author3, author4, autho5 = set(), set(), set(), set(), set()

for k, v in a1.items():
    if (v > 10):
        author1.add(k)
for k, v in a2.items():
    if (v > 10):
        author2.add(k)
for k, v in a3.items():
    if (v > 10):
        author3.add(k)
for k, v in a4.items():
    if (v > 10):
        author4.add(k)
for k, v in a5.items():
    if (v > 10):
        author5.add(k)
```

```python
unique1 = author1 - author2 - author3 - author4 - author5
unique2 = author2 - author1 - author3 - author4 - author5
unique3 = author3 - author1 - author2 - author4 - author5
unique4 = author4 - author1 - author2 - author3 - author5
unique5 = author5 - author1 - author2 - author3 - author4
```

```python
new_x = x.copy(deep = True)
new_y = y.copy(deep = True)

for i, r in new_x.iterrows():
    if (new_y.iloc[i]["author"] == "李白"):
        for c in r.items():
            if (c[0] in unique1):
                new_x.iloc[i][c[0]] = c[1] * 100
    elif (new_y.iloc[i]["author"] == "杜甫"):
        for c in r.items():
            if (c[0] in unique2):
                new_x.iloc[i][c[0]] = c[1] * 100
    elif (new_y.iloc[i]["author"] == "白居易"):
        for c in r.items():
            if (c[0] in unique3):
                new_x.iloc[i][c[0]] = c[1] * 100
    elif (new_y.iloc[i]["author"] == "蘇軾"):
        for c in r.items():
            if (c[0] in unique4):
                new_x.iloc[i][c[0]] = c[1] * 100
    elif (new_y.iloc[i]["author"] == "辛棄疾"):
        for c in r.items():
            if (c[0] in unique5):
                new_x.iloc[i][c[0]] = c[1] * 100
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(new_x, new_y, test_size = 0.1, stratify = new_y)
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier()
print("Training KNN...")
KNN.fit(x_train, y_train)
print("Predicting KNN...")
prediction = KNN.predict(x_test)
```

```python
result = y_test["author"].values
```

```python
n = 0

for i in range(len(prediction)):
    if (prediction[i] == result[i]):
        n += 1

print(f"Accuracy: {n / len(prediction) * 100}%")
```
